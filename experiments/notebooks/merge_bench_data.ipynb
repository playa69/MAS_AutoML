# Объединение данных benchmark

Объединяем `bench_merged.csv` с `bench_with_metrics.csv` для добавления метрик к данным benchmark.


```python
import pandas as pd
import numpy as np

# Загружаем данные
bench_merged = pd.read_csv('bench_merged.csv')
bench_with_metrics = pd.read_csv('bench_with_metrics.csv')

print(f"bench_merged: {bench_merged.shape}")
print(f"bench_with_metrics: {bench_with_metrics.shape}")
print("\nКолонки в bench_merged:")
print(bench_merged.columns.tolist())
print("\nКолонки в bench_with_metrics:")
print(bench_with_metrics.columns.tolist())
```

```python
# Просматриваем первые строки
print("Первые строки bench_merged:")
bench_merged.head()
```

```python
print("Первые строки bench_with_metrics:")
bench_with_metrics.head()
```

```python
# Определяем колонки метрик, которые нужно добавить
# Исключаем колонки, которые уже есть в bench_merged (task_id, task_name и метрики алгоритмов)
metrics_columns = ['dataset_size', 'num_features', 'num_cells', 'oof_score', 'train_score', 'test_score']

# Выбираем только нужные колонки из bench_with_metrics
metrics_to_join = bench_with_metrics[['task_id'] + metrics_columns].copy()

print("Колонки для объединения:")
print(metrics_to_join.columns.tolist())
print(f"\nКоличество строк с метриками: {len(metrics_to_join)}")
```

```python
# Объединяем данные по task_id (left join, чтобы сохранить все строки из bench_merged)
bench_final = bench_merged.merge(metrics_to_join, on='task_id', how='left')

print(f"Результат объединения: {bench_final.shape}")
print(f"\nКолонки в финальном датасете: {len(bench_final.columns)}")
print(bench_final.columns.tolist())
```

```python
# Проверяем результат
print("Первые строки объединенного датасета:")
bench_final.head(10)
```

```python
# Проверяем, сколько строк получили метрики
print(f"Всего строк: {len(bench_final)}")
print(f"Строк с метриками: {bench_final['dataset_size'].notna().sum()}")
print(f"Строк без метрик: {bench_final['dataset_size'].isna().sum()}")

# Показываем примеры строк с метриками и без
print("\nПримеры строк с метриками:")
bench_final[bench_final['dataset_size'].notna()][['task_id', 'task_name', 'dataset_size', 'num_features', 'num_cells']].head()
```

```python
b = bench_final[bench_final['dataset_size'].notna()]
```

```python
b
```

```python
import re

# Приводим dataset_size и num_features к int (если NaN - оставим NaN)
b['dataset_size'] = b['dataset_size'].apply(lambda x: int(x) if pd.notna(x) else x)
b['num_features'] = b['num_features'].apply(lambda x: int(x) if pd.notna(x) else x)

# Приводим oof_score к округлённому формату (4 знака)
if 'oof_score' in b.columns:
    b['oof_score'] = b['oof_score'].apply(lambda x: round(float(x), 4) if pd.notna(x) else x)

# Список колонок с моделями
model_cols = [
    'lightautoml','MLJAR_B','MLJAR_P','TPOT','constantpredictor','RandomForest','TunedRandomForest_star',
    'AutoGluon_B','AutoGluon_HQ','AutoGluon_HQIL','autosklearn','autosklearn2','flaml','GAMA_B','H2OAutoML'
]

def extract_score(cell_str):
    """Извлекает float из строки вида '0.993(0.002)2' -> 0.993"""
    if isinstance(cell_str, str):
        m = re.match(r'^([-\d\.]+)', cell_str)
        if m:
            try:
                return float(m.group(1))
            except Exception:
                return float('nan')
    elif isinstance(cell_str, (float, int)):
        return float(cell_str)
    return float('nan')

# Для каждой строки определим место lightautoml по oof_score среди моделей
places = []
for idx, row in b.iterrows():
    scores = []
    # Собрать пары (colname, score) только если score парсится нормально
    for col in model_cols:
        score = extract_score(row.get(col, float('nan')))
        scores.append((col, score))
    # Можно ли вычислить место? (lightautoml может быть nan иногда)
    lm_score = extract_score(row.get('lightautoml', float('nan')))
    # Сортируем по убыванию
    ranked = sorted(scores, key=lambda x: (-x[1], x[0]))
    place = next((i+1 for i, (col, score) in enumerate(ranked) if col == 'lightautoml'), None)
    places.append(place)

b['place'] = places

# Добавляем колонку: кто из моделей топ-1 и побил lightautoml (если да)
beaten_by = []
score_diff = []

for idx, row in b.iterrows():
    model_scores = []
    for col in model_cols:
        val = extract_score(row.get(col, float('nan')))
        model_scores.append((col, val))
    # Определяем топ-1
    sorted_models = sorted(model_scores, key=lambda x: (-x[1], x[0]))
    max_score = sorted_models[0][1]
    best_model = sorted_models[0][0]  # только 1 имя, даже если у других тот же скор
    # lightautoml score
    our_score = extract_score(row.get('lightautoml', float('nan')))
    # Добавим beaten_by только если топ-1 ≠ lightautoml И топ-1 > наш скор
    if not pd.isna(our_score) and (best_model != 'lightautoml') and (max_score > our_score):
        beaten_by.append(best_model)
    else:
        beaten_by.append("")
    # Разница с топ-1 (знак сохраняется: результат > 0 если lightautoml хуже, < 0 если лучше)
    if our_score == max_score:  # мы на первом месте
        # Второй по величине скор (может быть несколько с одинаковым баллом)
        next_scores = [val for _, val in sorted_models if (not pd.isna(val)) and (val < max_score)]
        if next_scores:
            score_gap_val = our_score - next_scores[0]
        else:
            score_gap_val = 0.0
        score_gap = f"{score_gap_val:+.3f}"
        score_diff.append(score_gap)
    else:
        score_gap_val = our_score - max_score
        score_gap = f"{score_gap_val:+.3f}"
        score_diff.append(score_gap)

b['beaten_by'] = beaten_by
b['score_gap'] = score_diff
```
